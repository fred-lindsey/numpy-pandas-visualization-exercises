1. Use matplotlib to plot the following equation:

    y = x^2 - x + 2

    You'll need to write the code that generates the x and y points.

    Add an anotation for the point 0, 0, the origin.

```python
x = range(-150, 300) #don't use []
# use a for loop, to loop through each element
y = [(num**2 - num + 2) for num in x]

plt.figure(figsize = (10,10))

plt.plot(x, y, c = 'red')
           
plt.axhline(y=0)
           
plt.axvline(x=0)

plt.text(0,0, s = 'origin', color = 'green', size = 16)

plt.xlim(-50, 100)

plt.ylim(-50, 400)
# %config InlineBackend.figure-format = 'png'
# plt.savefig('figure1.png')
```

```python
x = [x * 0.1 for x in range(0, 50)] #smoother curve when incr by 0.1
y1 = [(math.sqrt(num)) for num in x]
y2 = [(num**3) for num in x]
y3 = [(2**num) for num in x]
y4 = [(1/(num+1)) for num in x]

plt.figure(figsize = (8, 6))

plt.subplot(2,2,1)
plt.plot(x, y1)
plt.title('$y=sqrt(x)$')

plt.subplot(2,2,2)
plt.plot(x, y2)
plt.title('$y=x^3$')

plt.subplot(2,2,3)
plt.plot(x, y3)
plt.title('$y=2^x$')

plt.subplot(2,2,4)
plt.plot(x, y4)
plt.title('$y=1/(x+1)$')

#plt.xticks(range(0,16)) >> gives scaling increments
```
