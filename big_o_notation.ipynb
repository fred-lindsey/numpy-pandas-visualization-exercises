```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
x = range(1, 100)
y1 = [1 for n in x]
y2 = [math.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * math.log(n) for n in x]
y5 = [2**n for n in x]
y6 = [n**2 for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.plot(x, y1, c = 'blue', label = 'O1')
plt.plot(x, y2, c = 'green', label = 'O(log n)')
plt.plot(x, y3, c = 'yellow', label = 'O(n)')
plt.plot(x, y4, c = 'black', label = 'O(n log n)')
plt.plot(x, y5, c = 'pink', label = 'O(n^2)')
plt.plot(x, y6, c = 'brown', label = 'O(2^n)')
plt.plot(x, y7, c = 'magenta', label = 'O(n!)')
plt.plot(x, y8, c = 'purple', label = 'O(n^n)')
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.ylim(0, 5)
plt.xlim(0, 25)
plt.legend(loc = 'upper right')
```
