## Indexing and Subsetting

```python
import pandas as pd
import numpy as np

np.random.seed(123)

# Create list of values for names column.

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# Randomly generate arrays of scores for each student for each subject.
# Note that all the values need to have the same length here.

math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))


# Construct the DataFrame using the above lists and arrays.

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades,
                   'classroom': np.random.choice(['A', 'B'], len(students))})



df[['name', 'classroom']]
```

## .MERGE


```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
users.merge(roles, how="inner", left_on='role_id', right_on='id')

```

## EXERCISES 2

2. What is the result of using a right join on the DataFrames?

```python
(users.merge(roles, 
            how="right", 
            left_on='role_id', ##the 'join on' value, from priamry table
            right_on='id',
            indicator=True)
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id',
                    'name_x': 'employee',
                    'name_y': 'role',
                    'id_y': 'role_id'}
           )
)
                                                                      
## all roles returned, with a duplicate value. All left side
# roles returned
                                                                      
                                                                      
```

3. What is the result of using an outer join on the DataFrames?

```python
(users.merge(roles, 
            how="outer", 
            left_on='role_id', 
            right_on='id',
            indicator=True)
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id',
                    'name_x': 'employee',
                    'name_y': 'role',
                    'id_y': 'role_id'}
           )
)

##all users and roles are returned, along with nulls
```

7. Write the code necessary to create a cross tabulation of the number of titles by department. (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code to perform the manipulations.)

```python
#need: crosstab of titles by dept

#employees (emp_no) titles (emp_no) dept_emp (dept_no) department

emp_titles_depts = pd.read_sql("""
SELECT
    *
    FROM employees
    JOIN titles USING(emp_no)
    JOIN dept_emp USING(emp_no)
    JOIN departments USING(dept_no)
""", url)
emp_titles_depts
```

```python
pd.crosstab(emp_titles_depts.title, emp_titles_depts.dept_name, margins = True)
```
